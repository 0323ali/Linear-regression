```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('Salary_Data.csv')
a=data['YearsExperience'].values
b=data['Salary'].values
plt.scatter(a,b)
a_mean=np.mean(a)
b_mean=np.mean(b)
print('Mean of a={} and Mean of b={}'.format(round(a_mean,2),round(b_mean,2)))

m=len(a)
numi=0
denomi=0
for i in range(m):
  numi+=((a[i]-a_mean)*(b[i]-b_mean))
  denomi+=(a[i]-a_mean)**2
b1=numi/denomi
b0=b_mean-(b1*a_mean)
print('intercept is b0={} and slope is b1={}'.format(round(b0,2),round(b1,2)))
x=np.linspace(min(a),max(a))
y=b0+b1*x
plt.plot(x,y,color='red')

x_new=int(input('Enter the year of experience'))
y_new=b0+b1*x_new
print('Salary={}'.format(round(y_new,2)))
```
