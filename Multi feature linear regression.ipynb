```python
class colors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'

print(' ')
print(' ')
print(colors.FAIL+'----------------------------   Multi feature linear regrission   ----------------------------'+colors.ENDC)
print(' ')
print(' ')
print(' ')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('StudentsPerformance.csv')
math=data['math score'].values
reading=data['reading score'].values
writing=data['writing score'].values
plt.scatter(math,reading)
plt.xlabel('math score')
plt.ylabel('reading score')
plt.scatter(math,writing)
plt.xlabel('math score')
plt.ylabel('writing score')
plt.scatter(reading,writing)
plt.xlabel('reading score')
plt.ylabel('writing score')
m=len(reading)
x0=np.ones(m)
x=np.array([x0,math,reading]).T
y=np.array(writing)
b=np.array([0,0,0])
alpha=0.0001
def cost_function(x,y,b):
  m=len(y)
  j=np.sum((x.dot(b)-y)**2)/(2*m)
  return j
def gradient_descent(x,y,b,alpha,iterations):
  cost_history = [0] * iterations
  m=len(y)
  for i in range(iterations):
    h=x.dot(b)
    loss=h-y
    gradient=x.T.dot(loss)/m
    b = b-alpha*gradient
    cost = cost_function(x,y,b)
    cost_history[i] = cost
    return b, cost_history



newB, cost_history = gradient_descent(x, y, b, alpha, 100000)
print(colors.FAIL+'New values of B is: ',newB,colors.ENDC)
print(colors.FAIL+'Cost values is: ',cost_history[-1],colors.ENDC)
new_math=int(input(colors.OKGREEN +'Enter math score: ' + colors.ENDC))
new_reading = int(input(colors.OKGREEN + 'Enter reading score: ' + colors.ENDC))
s_writing=newB[0]+newB[1]*new_math+newB[2]*new_reading
print(colors.OKBLUE + 'predicting score of writing is: ',round(s_writing,1),colors.ENDC)
print(' ')
print(' ')
print(' ')
plt.scatter(math,reading)
plt.xlabel('math score')
plt.ylabel('reading score')
plt.scatter(math,writing)
plt.xlabel('math score')
plt.ylabel('writing score')
plt.scatter(new_math,new_reading,color='red')
```
